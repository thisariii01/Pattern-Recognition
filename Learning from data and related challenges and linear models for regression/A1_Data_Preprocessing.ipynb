```python
import numpy as np
import matplotlib.pyplot as plt

def generate_signal (signal_length,num_nonzero):
    signal = np.zeros(signal_length)
    nonzero_indices = np.random.choice(signal_length,num_nonzero,replace = False)
    nonzero_values = 50 * np.random.randn(num_nonzero)
    signal[nonzero_indices]= nonzero_values
    return signal

# Data generation
signal_length = 100 # Total length of the signal
num_nonzero = 10
your_index_no = 200023 # Index number ( without English letter and without leading zeros )
signal = generate_signal(signal_length,num_nonzero)
signal[10] = (your_index_no % 10) * 10 + 10
if your_index_no % 10 == 0:
    signal[10] = np. random.randn(1) + 30

signal = signal.reshape(signal_length ,1)

# plotting
plt.figure(figsize =(15 ,5))
plt.subplot(1, 1, 1)
plt.title(" Data ")
plt.stem(signal)
```

```python
# MaxAbs Scaler
from sklearn import preprocessing
max_abs_scaler = preprocessing.MaxAbsScaler()
scaled_signal_MaxAbs = max_abs_scaler.fit_transform(signal)
```

```python
# Min-Max scaling
x_min = min(signal)
x_max = max(signal)
scaled_signal_MinMax = []
for x in signal:
  x_scaled = (x - x_min) / (x_max - x_min)
  scaled_signal_MinMax.append(x_scaled)
```

```python
# Standardization scaling
x_mean = np.mean(signal)
x_std = np.std(signal)
scaled_signal_Standardization = []
for x in signal:
  x_scaled = (x - x_mean) / x_std
  scaled_signal_Standardization.append(x_scaled)
```

```python
# Visulaizing
fig = plt.figure(figsize =(16 ,20))
ax1 = fig.add_subplot(411)
ax2 = fig.add_subplot(412)
ax3 = fig.add_subplot(413)
ax4 = fig.add_subplot(414)
ax1.stem(signal)
ax1.set_title("Original data")
ax2.stem(scaled_signal_MaxAbs)
ax2.set_title("Max-Abs scaled data")
ax3.stem(scaled_signal_MinMax)
ax3.set_title("Min-Max scaled data")
ax4.stem(scaled_signal_Standardization)
ax4.set_title("Standardization scaled data")
plt.show()
```
